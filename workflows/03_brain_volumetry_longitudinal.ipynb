# 03 — Longitudinal brain volume change with SIENA

**SIENA** (FSL) measures percentage brain volume change (PBVC) between two timepoints — a key MS biomarker (annualised brain atrophy ≈ 0.4% in healthy adults; ≥0.5–1% suggests pathological atrophy).

Inputs:
- `T1_tp1.nii.gz` — baseline
- `T1_tp2.nii.gz` — follow-up (≥6 months apart, ideally same scanner/sequence)

```python
from pathlib import Path
DATA = Path('/neurodesktop-storage/ms-workshop-data/sub-001-longitudinal')
OUT  = Path('/neurodesktop-storage/ms-workshop-out/sub-001/siena')
OUT.mkdir(parents=True, exist_ok=True)
TP1 = DATA / 'sub-001_ses-01_T1w.nii.gz'
TP2 = DATA / 'sub-001_ses-02_T1w.nii.gz'
print(TP1.exists(), TP2.exists())
```

```python
%%bash -s "$TP1" "$TP2" "$OUT"
ml load fsl
TP1=$1; TP2=$2; OUT=$3
siena $TP1 $TP2 -o $OUT
```

SIENA outputs land in `$OUT/`:
- `report.siena` — text summary including **PBVC** (percent brain volume change)
- HTML report with images of the edge motion analysis

```python
report = (OUT / 'report.siena').read_text()
print(report)
```

## Discussion

- **Annualised PBVC** is the headline metric in MS atrophy studies and DMT trials.
- **Pseudoatrophy** in early treatment (resolution of inflammatory oedema) can confound short-interval scans — use ≥1 year follow-up where possible.
- Cross-sectional alternative: **SIENAX** (single timepoint, normalised brain volume).
- More sophisticated: **FreeSurfer's longitudinal pipeline** for cortical/subcortical volumes.
